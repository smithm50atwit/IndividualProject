```python
import matplotlib.pyplot as plt

years = [2023, 2024, 2025, 2026, 2027]
values = [211, 245.25, 279.5, 314.75, 350]  

plt.figure(figsize=(8, 5))
plt.plot(years, values, marker='o', linestyle='-', color='green')
plt.title('Projected Growth of Global Second-Hand Market (2023–2027)')
plt.xlabel('Year')
plt.ylabel('Market Value (Billion USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
years = [2022, 2023, 2024, 2025, 2026, 2027]
market_share = [12, 15, 17.5, 20, 22.5, 25] 


plt.figure(figsize=(8, 5))
plt.bar(years, market_share, color='steelblue')
plt.title('U.S. Second-Hand Apparel Market Share (%)')
plt.xlabel('Year')
plt.ylabel('Market Share (%)')
plt.ylim(0, 30)
plt.tight_layout()
plt.show()
```

```python
platforms = ['eBay', 'Goodwill/Savers', 'StockX', 'GOAT', 'ThredUp']
estimated_popularity = [80, 65, 90, 85, 70]  

plt.figure(figsize=(9, 5))
plt.barh(platforms, estimated_popularity, color='purple')
plt.xlabel('Popularity Index')
plt.title('Resale Platform Popularity for Shoes & Apparel')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
stockx_prices = [310, 320, 305, 330, 350, 360, 355, 370, 365, 380, 375, 390]
retail_price = [200] * len(months)  

plt.figure(figsize=(10, 6))
plt.plot(months, stockx_prices, marker='o', label='StockX Resale Price', color='green', linewidth=2)
plt.plot(months, retail_price, linestyle='--', label='Retail Price ($200)', color='gray', linewidth=2)


plt.title('StockX Price Trend vs. Retail Price for Air Jordans (2024)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(150, 420)
plt.tight_layout()

plt.show()
```
